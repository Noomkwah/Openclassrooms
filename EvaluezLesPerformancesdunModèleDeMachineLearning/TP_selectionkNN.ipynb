```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import neighbors, metrics
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
```

```python
data = pd.read_csv("winequality_white.csv", sep=";")
data.head()
```

On va extraire les données, en séparant d'un côté les points et d'un autre les étiquettes.

```python
y = data["quality"].values
X = data[data.columns[:-1]].values
```

Affichons maintenant un histogramme de chacune de nos variables.

```python
fig = plt.figure(figsize=(16,12))

for feat_idx in range(X.shape[1]):
    ax = fig.add_subplot(3,4, (feat_idx+1))
    h = ax.hist(X[:, feat_idx], bins=50, color='steelblue', density=True, edgecolor='none')
    ax.set_title(data.columns[feat_idx], fontsize=14)
plt.show()
```

On remarque que les variables prennent leurs valeurs dans des ensembles différents. Par exemple, alcohol varie de 8 à 14, tandis que pH varie de 2.75 à 3.75. Il nous faut donc standardiser les données.

```python
y_class = np.where(y<6, 0, 1) # On commence par séparer les qualités selon qu'elles soient médiocres(0) ou bonnes (1).
y_class
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y_class, test_size = 0.3)
```

```python
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
fig = plt.figure(figsize=(16, 12))
for feat_idx in range(X_train_scaled.shape[1]):
    ax = fig.add_subplot(3,4, (feat_idx+1))
    h = ax.hist(X_train_scaled[:, feat_idx], bins=50, color = 'steelblue', density=True, edgecolor='none')
    ax.set_title(data.columns[feat_idx], fontsize=14)
plt.show()
```

Notre but, c'est de trouver quel k prendre pour réaliser un k-NN. On va faire une validation croisée.

```python
# Fixons les valeurs de l'hyper-paramètre que l'on veut tester.
param_grid = {'n_neighbors':[3,5,7,9,11,13,15]}

# On choisit le score que l'on veut tester, ici l'accuracy, c'est-à-dire la proportion de prédictions correctes.
score = "accuracy"

# On crée un classifieur kNN avec recherche de l'hyper-paramètre par validations croisée
clf = model_selection.GridSearchCV(neighbors.KNeighborsClassifier(), # un classifieur kNN
                                  param_grid, # Hyperparamètres à tester
                                  cv=5, # Nombre de folds de la validation croisée
                                  scoring=score) # score à optimiser

clf.fit(X_train_scaled, y_train)

# On affiche le(s) hyper-paramètre(s) optimal(ux)
print("Meilleur(s) hyperparamètre(s) sur le jeu d'entraînement:")
print(clf.best_params_)

# Afficher les performances correspondantes
print("Résultats de la validation croisée :")
for mean, std, params in zip(
        clf.cv_results_['mean_test_score'], # score moyen
        clf.cv_results_['std_test_score'],  # écart-type du score
        clf.cv_results_['params']           # valeur de l'hyperparamètre
    ):

    print("{} = {:.3f} (+/-{:.03f}) for {}".format(
        score,
        mean,
        std*2,
        params
    ) )
```

On peut maintenant regarder la performance sur le jeu de test. GridSearchCV a automatiquement réentrainé le meilleur modèle sur l'intégralité du jeu d'entrainement.

```python
y_pred = clf.predict(X_test_scaled)
print("\nSur le jeu de test : {:.3f}".format(metrics.accuracy_score(y_test, y_pred)))
```
