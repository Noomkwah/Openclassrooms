```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("my_courses.csv", index_col=0)
data.head()
```

```python
# PARTIE 1 - NETTOYAGE DES DONNÉES
data.dropna(inplace=True) # On supprime les lignes avec des éléments manquants.
data.drop(columns=["idCours"], inplace=True) # L'id du cours ne nous donne aucune information sur laquelle faire une ACP, on supprime.
# Il n'y a pas de raison qu'il y ait d'outliers ou de doublons.
```

```python
# PARTIE 2 - ACP

# On commence par centrer-réduire nos données.
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
data_scaled = pd.DataFrame(data_scaled, index=data.index, columns = data.columns)
data_scaled.head()
```

```python
data_scaled.describe()
```

```python
# On peut maintenant réaliser l'ACP.
pca = PCA()
data_projected = pca.fit_transform(data_scaled)
data_projected = pd.DataFrame(data_projected, index = data_scaled.index, columns =  ["PC%u" % (i+1) for i in range(9)])
data_projected.head()
```

```python
# pca.explained_variance_ratio_ nous donne la variance expliquée par chaque composantes principale.

# On trace l'éboulis des valeurs propres
plt.plot(range(len(pca.explained_variance_ratio_)), pca.explained_variance_ratio_*100, marker="o", label="Variance")

# On trace la variance cumulée expliquée par les k premières composantes principales
plt.plot(range(len(pca.explained_variance_ratio_)), np.cumsum(pca.explained_variance_ratio_)*100, linestyle='-', color='red', label="Variance cumulée")

# On rend le graphique plus joli
plt.xticks(range(9), ["PC%u" % u for u in range(1,10)])
plt.grid()
plt.title("Éboulis des valeurs propres")
plt.xlabel("Composantes principales")
plt.ylabel("Variance expliquée (%)")
plt.legend()
plt.show()
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
# On voit que la 4ème composante explique moins de 10% de la variance, 
# c'est-à-dire moins que 100%/p (où p=10 est le nombre de composantes)
# On va donc ne considérer que les trois premières composantes, qui ensemble expliquent 74.93% de la variance.
# data_projected[["PC1", "PC2", "PC3"]]

data_scaled = data_scaled.merge(data_projected[["PC1", "PC2", "PC3"]], how="inner", left_index=True, right_index=True)
data_scaled.head()
```

```python
# On va maintenant examiner les corrélations entre les variables et les composantes principales.
sns.heatmap(data_scaled.corr().iloc[:-3, -3:], annot=True, cmap="coolwarm")
plt.show()
```

Sur la matrice de corrélation ci-dessus, on voit que la PC1 est fortement corrélée à la durée du cours, à son nombre de chapitre, et au nombre d'évaluation. De même, elle est fortement corrélée négativement à la progression. On peut donc dire que la composante PC1 correspond en quelque sorte à la "longueur du cours".

La PC2 est fortement corrélée à la date de dernière mise à jour, à la difficulté et est négativement corrélée à la moyenne de classe. Autrement dit, on peut dire que la composante PC2 correspond en quelque sorte à la "difficulté constatée du cours".

La PC3 est fortement corrélée négativement au ratioQuizEvaluation, et corrélé positivement à la moyenne de classe. Autrement dit, d'après ChatGPT, cette composante PC3 correspond à la "qualité pédagogique du cours".
