## 2 - Préparation des données et classifieur naïf

### Préparation des données

On commence par séparer notre jeu en features et labels, puis en un jeu d'entraînement et un jeu de test.

```python
X = data.drop(columns=["target"])
y = data["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
X_train.shape
```

### Premières modélisations

On va entraîner un classifieur naïf, qui prédit toujours la classe la plus fréquente observée dans la base de données.

```python
from sklearn.dummy import DummyClassifier
dum = DummyClassifier(strategy = "most_frequent")
dum.fit(X_train, y_train)
```

```python
y_dummy = dum.predict(X_test)
```

```python
# On sera sûrement amené à réutiliser ce code, donc autant le mettre dans une fonction.
def score(estimator):
    """Returns the accuracy score on training AND on test set."""
    train_score = estimator.score(X_train, y_train).round(4)
    test_score = estimator.score(X_test, y_test).round(4)
    print(f"Score train :  {train_score}\nScore test : {test_score}")
```

```python
score(dum)
```

Sans surprise, le classifieur naïf a en fait reproduit dans les proportions de la base de données les catégories observées.

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_dummy)
mat = pd.DataFrame(mat)
mat
```

```python
mat.columns = [f"pred_{i}" for i in mat.columns]
mat.index = [f"test_{i}" for i in mat.index]
mat
```

```python
# Mettons le code précédent dans une fonction pour le rendre plus facile d'utilisation.
def confusion(y_test, y_pred):
    """Display a fancy confusion matrix"""
    mat = confusion_matrix(y_test, y_pred)
    mat = pd.DataFrame(mat)
    mat.columns = [f"pred_{i}" for i in mat.columns]
    mat.index = [f"test_{i}" for i in mat.index]
    
    return mat
```

```python
confusion(y_test, y_dummy)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, threshold = roc_curve(y_test, y_dummy)
roc_auc = auc(false_positive_rate, true_positive_rate)
roc_auc
```

```python
score(log_regr)
```

C'est assez étonnant d'avoir un score test aussi bon que le score train. Mais c'est normal, ce dataset est didactique.

```python
confusion(y_test, y_log)
```

Pour obtenir les taux de faux positifs, vraies positifs etc. on utilise ce code :

```python
def rates(y_test, y_pred):
    """returns (TP, FP, TN, FN) which stands for:
    TP : False Positives
    FP : False Negatives
    TN : True Positives
    FN : True Negatives"""
    mat = confusion_matrix(y_test, y_pred)
    FP = (mat.sum(axis=0) - np.diag(mat))[0]
    FN = (mat.sum(axis=1) - np.diag(mat))[0]
    TP = np.diag(mat)[0]
    TN = mat.sum() - (FP + FN + TP)
    
    return (TP, FP, TN, FN)
```

```python
rates(y_test, y_log)
```

```python
def various_scores(y_test, y_pred):
    TP, FP, TN, FN = rates(y_test, y_pred)
    
    recall = (TP/(TP+FN)).round(4)
    specificity = (TN/(TN+FP)).round(4)
    precision = (TP/(TP+FP)).round(4)
    
    
    print(f"recall = {recall}\nspecificity = {specificity}\nprecision = {precision}")
```

```python
various_scores(y_test, y_log)
```

## 4 - Optimisation des modèles

On va optimiser les modèles en choisissant, par validation croisée, le bon hyperparamètre.

```python
from sklearn.model_selection import GridSearchCV
```

```python
estimator = LogisticRegression()
params = {
    'C' : np.logspace(-3, 3, 7),
    'penalty' : ["l1", "l2"],
    'solver' : ["newton-cg", "lbfgs", "liblinear"]
}
```

```python
grid = GridSearchCV(estimator, 
                   params,
                   cv = 10, # nombre de partitions lors de l'entraînement
                   n_jobs=-1, # cela force le gridsearch à travailler sur tous les CPU de notre machine
                   return_train_score=True,
                   verbose=1)
grid.fit(X_train, y_train)
```

```python
best_params = grid.best_params_
best_params
```

```python
res = grid.cv_results_
res = pd.DataFrame(res)
res
```

```python
cols = [i for i in res.columns if "split" not in i]
res = res[cols].sort_values("rank_test_score")
res.head(10)
```

```python
def resultize(grid):
    res = grid.cv_results_
    res = pd.DataFrame(res)
    cols = [i for i in res.columns if "split" not in i]
    res = res[cols].sort_values("rank_test_score")
    
    return res    
```

```python
best_estimator = LogisticRegression(**best_params)
best_estimator.fit(X_train, y_train)
y_pred = best_estimator.predict(X_test)
```

```python
score(best_estimator)
```

```python
confusion(y_pred, y_test)
```

```python
rates(y_pred, y_test)
```

```python
estimator = LinearSVC()
params = {"C" : np.logspace(-3,3,7)}
grid = GridSearchCV(estimator,
                   params,
                   cv=10,
                   n_jobs=-1,
                   return_train_score=True,
                   verbose=1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
resultize(grid)
```

```python
best_svc = LinearSVC(**grid.best_params_)
best_svc.fit(X_train, y_train)
y_pred_svc = best_svc.predict(X_test)
```

```python
score(best_svc)
```

```python
confusion(y_pred_svc, y_test)
```

```python
rates(y_pred_svc, y_test)
```

```python
various_scores(y_pred_svc, y_test)
```
